# Simple Aleatory Sampling
Cada atleta té un ID únic, genera n nº aleatoris entre 1 i el nº d'atletes, i selecciona aquells que coincideix amb l'ID

```python
def SAS(n,df): #n és el tamany de la mostra
    import random
    import pandas as pd
    df_agrupat=df.groupby('ID')
    N=df_agrupat.ngroups
    llista=[]
    sample={}
    for i in range(n):
        llista.append(random.randint(1, N))
    llista.sort()
    for key, item in df_agrupat:
        if key in llista:
            sample[key]=item
    return sample
```

S'obté un diccionari amb key: ID del atleta, i value: Dataframe de totes les vegades que ha participat en els jocs olímpics

```python
diccionari_samples=SAS(5,df)
```

```python
for i in diccionari_samples:
    print(i)
    print(diccionari_samples[i])
    print('\n\n')
```

# Systematic Sampling

```python
def SS(n,df): #n és sample size
    import pandas as pd
    import math
    df_agrupat=df.groupby('ID')
    N=df_agrupat.ngroups
    num_mostres=math.floor(N/n) #S'arrodoneix cap abaix per si hi ha decimals, si s'arrodoneix cap amunt podria no agafar-se el sample size indicat
    llista=[]
    sample={}
    for i in range(num_mostres,N+1,num_mostres):
        llista.append(i)
    for key, value in df_agrupat:
        if key in llista:
            sample[key]=value
    return sample
        
```

```python
diccionari_samples2=SS(6,df)
```

```python
for i in diccionari_samples2:
    print(diccionari_samples2[i])
    print('\n\n')
```

# Stratified Sampling
Primer separar els estrats, i assignar nous IDs

```python
df_sexe=df.groupby('Sex')

gb_estacio=df.groupby('Season')

df_hivern=pd.DataFrame(gb_estacio.get_group('Winter'))
df_hivern['ID'] = df_hivern.groupby('ID').ngroup()+1
gb_hivern=df_hivern.reset_index(drop=True).groupby('ID')

df_estiu=pd.DataFrame(gb_estacio.get_group('Summer'))
df_estiu['ID'] = df_estiu.groupby('ID').ngroup()+1
gb_estiu=df_estiu.reset_index(drop=True).groupby('ID')


print(gb_estacio.size())
print(gb_hivern.get_group(1))
print(gb_estiu.get_group(1))
```

2n utilitzar el Simple Aleatory Sampling per cada estrat per seleccionar les mostres (n mostres en cada estrat)

```python
diccionari_samples_hivern=SAS(3,df_hivern)
diccionari_samples_estiu=SAS(3,df_estiu)
```

```python
print('HIVERN')
for i in diccionari_samples_hivern:
    print(diccionari_samples_hivern[i])
    print('\n\n')
    
print('ESTIU')
for i in diccionari_samples_estiu:
    print(diccionari_samples_estiu[i])
    print('\n\n')
```

# Reservoir sampling

```python
def reservoir(df,n):
    import pandas as pd
    import random
    stream= df.groupby('ID')
    reservoir=[]
    stream.get_group(1)
    N=len(stream)
    for x in range(1,n+1):
        reservoir.append(stream.get_group(x))    
    for i in range(n+2,N):
        j = random.randrange(0,i+1)
        if(j < n):
            reservoir[j]=stream.get_group(i)
    return reservoir
```

```python
reservoir_samples=reservoir(df,3)
```

```python
for i in reservoir_samples:
    print(i)
```
